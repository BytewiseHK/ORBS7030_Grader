## 3.1 If different flight distance related to customers' satisfaction?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Airline_satisfaction.csv')

satisfaction_mapping = {'satisfied': 10, 'neutral or dissatisfied': 1}
df['satisfaction_numeric'] = df['satisfaction'].map(satisfaction_mapping)

plt.figure(figsize = (5, 6))
plt.scatter(df['satisfaction_numeric'], df['Flight Distance'], color = 'b', alpha = 0.6)

plt.title('Relationship Between Satisfaction and Flight Distance')
plt.xlabel('Satisfaction (0: Neutral/Dissatisfied, 10: Satisfied)')
plt.ylabel('Flight Distance')
plt.grid(True)

plt.show()
```

```python
import pandas as pd

df = pd.read_csv('Airline_satisfaction.csv')

satisfaction_mapping = {'satisfied': 10, 'neutral or dissatisfied': 1}
df['satisfaction_numeric'] = df['satisfaction'].map(satisfaction_mapping)

correlation = df['satisfaction_numeric'].corr(df['Flight Distance'])

print(f"pearson: {correlation:.4f}")
```

From the scatter diagram it illustrates that there is not a significant correlation between Flight distance and customer satisfaction. Moreover, the pearson Pearson correlation coefficient close to 0, so it also shows that there is a weak positive correlation among these tow variables. So different flight distance minimal related to customers' satisfaction.

# 4. Hypothesis Testing

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats 
from statsmodels.stats.weightstats import ztest

order = pd.read_csv('Airline_satisfaction.csv')

print(order.nunique())

print(order.describe())

order.groupby('Customer Type')['Departure Delay in Minutes'].mean()
```

$H_0$: The average time of Departure Delay in Minutes to loyal customer is 15 minutes.  
$H_1$: The average time of Departure Delay in Minutes to loyal customer is NOT equal to 15 minutes.

```python
from statsmodels.stats.weightstats import ztest
DD_in_Minutes_loyal = order[order['Customer Type'] == 'Loyal Customer']['Departure Delay in Minutes']
print(DD_in_Minutes_loyal)
z_stat, p_value = ztest(DD_in_Minutes_loyal, value = 15)
print(f"Z-statistic: {z_stat:.4f}, p-value: {p_value:.4f}")
```

Set $\alpha = 0.05$. Given p-value is greater than 0.05, we fail to reject the $H_0$ and conclude that The average time of Departure Delay in Minutes to loyal customer is not significantly different from 15 minutes.
